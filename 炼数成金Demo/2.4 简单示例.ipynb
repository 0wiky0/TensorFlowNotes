```python
import tensorflow as tf
import numpy as np

# 生成100个随机样本点
x_data = np.random.rand(100)
y_data = x_data * 0.1 + 2
print(x_data)
print(y_data)
# 构造线性模型
b = tf.Variable(0.0)
k = tf.Variable(0.0)
y = k*x_data + b

# tf.square(y_data - y) 平方差 
# reduce_mean 求平均值 
# y_data -> 真实值 ， y -> 预测值
loss = tf.reduce_mean(tf.square(y_data - y))

# 定义一个梯度下降法来进行训练的优化器
optimizer =  tf.train.GradientDescentOptimizer(0.2)

# 训练使loss"最小",即使预测出来的k和b最接近真实值
train = optimizer.minimize(loss)

# 初始化全部变量
init = tf.global_variables_initializer()

# 常用写法（无需手动关闭会话）
with tf.Session() as sess:
    # 执行以上的op
    sess.run(init)
    # 循环自增
    for step in range(401):
        sess.run(train)
        if step % 20 == 0:
            print(step,sess.run([k,b]))
```
